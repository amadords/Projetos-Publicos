**O que é COVID-19?**

É uma doença que provém de uma família de vírus muito comum em diferentes espécies de animais, a exemplo de camelos, gado, gatos e morcegos.

De uma forma um tanto rara, os coronavírus que infectam animais podem ultrapassar a barreiras das espécies e infectar pessoas, como exemplo do MERS-CoV e SARS-CoV e Recentemente,o SARS-CoV-2, doença que teve os primeiros relatos confirmado na China em 2019 e, posteriormente passou a nível de Epidemia e hoje, Pandemia.

O quadro clínico de pessoas infectadas pelo vírus é bastante variado, indo de infecções assintomáticas a quadros graves. De acordo com a Organização Mundial de Saúde, a maioria (cerca de 80%) dos pacientes com COVID-19 podem ser assintomáticos ou oligossintomáticos (poucos sintomas), e aproximadamente 20% dos casos detectados requer atendimento hospitalar por apresentarem dificuldade respiratória, dos quais aproximadamente 5% podem necessitar de suporte ventilatório, como informa a [página oficial no Brasil](https://coronavirus.saude.gov.br) sobre o vírus.

 
**Quais são os sintomas?**

Os sintomas da COVID-19 podem variar de um resfriado, a uma Síndrome Gripal-SG (presença de um quadro respiratório agudo, caracterizado por, pelo menos dois dos seguintes sintomas: sensação febril ou febre associada a dor de garganta, dor de cabeça, tosse, coriza) até uma pneumonia severa. Sendo os sintomas mais comuns:

- Tosse
- Febre
- Coriza
- Dor de garganta
- Dificuldade para respirar
- Perda de olfato (anosmia)
- Alteração do paladar (ageusia)
- Distúrbios gastrintestinais (náuseas/vômitos/diarreia)
- Cansaço (astenia)
- Diminuição do apetite (hiporexia)
- Dispnéia ( falta de ar)

 
**Como é transmitido?**

A transmissão acontece de uma pessoa doente para outra ou por contato próximo por meio de:

- Toque do aperto de mão contaminadas;
- Gotículas de saliva;
- Espirro;
- Tosse;
- Catarro;
- Objetos ou superfícies contaminadas, como celulares, mesas, talheres, maçanetas, brinquedos, teclados de computador etc.

 
Diagnóstico

O diagnóstico da COVID-19 pode ser realizado a partir diagnóstico clínico, diagnóstico clínico-imagem, diagnóstico laboratorial, diagnóstico laboratorial em indivíduos assintomáticos, cada um seguindo determinado critério.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pprint import pprint

sns.set_style()
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

**Colocando os valores "extras" em uma variáves a parte**

```
df_totais =  df.query("location in ['Africa', 'Europe', 'European Union', 'North America', 'Oceania', 'South America', 'Asia', 'World']")

df_totais.location.unique()
```

### **Quais países têm mais casos?**

> Veja que até 25 de fevereiro de 2021 os países com mais casos no mundo são **Estados Unidos**, **India**, **Brasil**, **Reino Unido** e **Russia**, respectivamente.

```
df_clean.loc[df.date == '2021-02-25'].sort_values(by="total_cases", ascending=False)
```

### **Quais países têm mais mortes?**

#### **Gráfico em escalas linear e logarítmica**

> A **Escala Linear** ou aritmética é comum e rotineiramente utilizada até pelas pessoas mais leigas, plotando a variável e `X` e a correspondente em `y`. 

> A **Escala Logarítmica** pode ser contruída a partir de uma multiplicação,  a exemplo de 10 ou potências de 10, não assumindo valores negativos. A vantagem é que ela utiliza o logaritmo de uma grandeza ao invés da própria grandeza, o que pode ser útil principalmente para a visualização de grandes quantidades de dados. Utilizada em vários contextos,sua função é dada por:

$$
log_a b = x \Leftrightarrow a^x = b
$$

> 




```
sns.set()
# criando variáveis que serão utilizadas
tamanho = list(range(366))
casos = df_portugal['total_cases']
mortes = df_portugal['total_deaths']

# gráfico em escalar linear
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(9, 7))
ax1.set_title('Casos em Portugal (Escala linear)')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(tamanho, casos)

# gráfico em escala logarítmica

ax2.set_title('Casos em Portugal (Escala logarítmica)')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(tamanho, casos)

ax3.set_title('Mortes em Portugal')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Mortes')
ax3.plot(tamanho, mortes)

ax4.set_title('Mortes em Portugal (Escala logarítmica')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(tamanho, mortes)
plt.tight_layout()
```

**Setando a data para o índice**

```
portugal_temporal = df_portugal.set_index('date')
portugal_temporal
```

#### **Novos Casos confirmados em Portugal - Gráfico**

```
fig, ax = plt.subplots()

portugal_temporal["new_cases"].plot(kind="line", ax=ax, figsize=(15, 6))

ax.set_title("Novos Casos")
ax.set_ylabel("Total")
ax.set_xlabel("Data")


plt.tight_layout()
```

**Veja**

> Os dados de casos e mortes têm valores muito destoantes, por isso não podemos plotar os gráficos das duas em um só gráfico, a menos que reescalemos os dados, ou seja, os coloquemos em uma mesma escala.

```
print("O número máximo de novos casos em um dia é: ", portugal_temporal['new_cases'].max())
print("O número máximo de novas mortes em um dia é: ", portugal_temporal['new_deaths'].max())
```

#### **Número de Novos Casos e Novas Mortes - Gráfico**

```
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(15,6))

# plotar temperatura média
portugal_temporal["new_cases"].plot(kind="line", ax=ax1)

ax1.set_title("Gráfico de Novos Casos")
ax1.set_ylabel("Novos Casos")
ax1.set_xlabel("Data")

# plotar casos confirmados
portugal_temporal["new_deaths"].plot(kind="line", ax=ax2)
ax2.set_title("Gráfico de Novas Mortes")
ax2.set_ylabel("Novas Mortes")
ax2.set_xlabel("Data")


plt.tight_layout()
```

#### **Reescalando os dados**

> Dessa forma consiguiremos plotar os gráficos juntos.

```
# criando variável com as variáveis que queremos

df_norm = portugal_temporal[['new_cases', 'new_deaths']]

# novos dados
df_norm.head()
```

**Normalizando ou reescalando**

```
df_norm = (df_norm - df_norm .mean()) / df_norm .std()
```

**Dados normalizados**

```
df_norm.head()
```

**Últimos dados**

```
df_norm.tail()
```

**Visualizando onde inicia os dados**

```
pprint(df_norm.loc['2020-03-16'])

pprint(df_norm.loc['2020-03-17'])
```

#### **Gráfico com Dados Normalizados**

```
# plotar os gráficos normalizados
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,6))

# novos casos
df_norm.loc['2020-03-17':'2021-02-11', 'new_cases'].plot(kind="line", ax=ax)

# novas mortes
df_norm.loc['2020-03-17':'2021-02-11', 'new_deaths'].plot(kind="line", ax=ax)


# eixo-x e title
ax.set_xlabel("Data")
ax.set_title("Novos Casos Confirmado x Novas Mortes")

plt.tight_layout()
```

> Veja como a perceptível o quanto o número de mortes e novos casos andam juntos, em relação à oscilação.

**Regressão**
> A correlação é tão visível que facilmente plotamos uma linha de regressão linear.

```
fig, ax = plt.subplots( figsize=(15,8))


df_norm.plot.scatter(x = 'new_cases', y = 'new_deaths', ax=ax, s=5)

sns.regplot('new_cases', 'new_deaths', scatter_kws={"s": 0.2},
            line_kws={"color": "red"}, data = df_norm, ax=ax)

plt.show();
```

**Outra visualização da mesma correlação**

```
%matplotlib inline

sns.lmplot(x="new_cases", y="new_deaths", palette="Pastel1", size = 12, data = df_norm);
```

#### **Correlação entre Novos Casos e Mortes**

**Portugal**
> Veja que em Portugal, bem como foi no [Brasil](https://github.com/amadords/Projetos-Publicos/blob/master/Panorama_da_Covid_19_no_Brasil.ipynb) e também  como conseguimos ver em todo o mundo, a correlação entre novos casos e morte é altíssima.

#### **Total de Casos e Mortes por Dia do Mês**
> A intenção é saber quais dias, no total, ocorrem mais e menos novos casos e novas mortes.

**Confirmando que os 31 dias estão presentes**

```
portugal_temporal.index.day.unique().sort_values(ascending = True)
```

**Quais os 5 dias com mais novos casos?**

```
mais_casos = portugal_temporal['new_cases'].groupby(portugal_temporal.index.day).sum().sort_values(ascending = False).head()
mais_casos
```

**Quais os 5 dias com mais novas mortes?**

```
mais_mortes = portugal_temporal['new_deaths'].groupby(portugal_temporal.index.day).sum().sort_values(ascending = False).head()
mais_mortes
```

**Quais os 5 dias com menos novos casos?**

```
menos_casos = portugal_temporal['new_cases'].groupby(portugal_temporal.index.day).sum().sort_values(ascending = False).tail()
menos_casos
```

**Quais os 5 dias com menos novas mortes?**

```
menos_mortes = portugal_temporal['new_deaths'].groupby(portugal_temporal.index.day).sum().sort_values(ascending = False).tail()
menos_mortes
```
