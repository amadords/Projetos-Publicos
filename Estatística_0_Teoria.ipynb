"*Conjunto de técnicas que permite, de forma sistemática, organizar, descrever, analisar e interpretar dados oriundos de estudos ou experimentos, realizados em qualquer área do conhecimento." (Antonio Carlos P. de Lima, 2015)* 

- É utilizada em Machine Learning, Análise de Dados etc.
- De forma geral é dividida em três grande áreas:
  - Descritiva
  - Probabilística
  - Inferencial

**1. Descritiva**
- Organiza, demonstra e resume os dados.

**2. Probabilística**
- Analisa situações sujeitas ao acaso.

**3. Inferência**
- Busca obter respostas sobre um fenômeno com dados representativos, a partir de amostras.

## Conceitos importantes
**- Observação e Experimento:**
  - Observação: Estudo em que os elementos analisados não são afetados (pesquisa).
  - Experimento: Condições ou tratamentos são impostar a grupos, para avaliar o resultado.
    - Divido em grupos: A, B, C etc. e grupo de controle (placebo). Nesses grupos são avaliadas questões de causa e efeito.

**- Variáveis:**
  - Quantitativa (Numéricas)
    - Contínuas: Valores reais. Podem assumir qualquer intervalo.
    - Discretas: Números fixos. Deve ser inteiro e em um intervalo.
  - Qualitativa (Categóricas)
    - Nominais: Categorias SEM hierarquia.
    - Ordinais: Categorias COM hierarquia.

O objetivo é obter informações ocultas sobre os dados.
  - Variação
  - Anomalias
  - Distribuição
  - Tendências
  - Padrões
  - Relações

Tomada de decisão baseada em dados passa pela EDA, para depois buscar solução para o problema.

# Amostragem

- A amostra é parte de uma população, selecionada usando alguma técnica que dê chances iguais de serem selecionadas aos elementos da população 

### Conceitos

- **População:** Alvo do estudo. Público eleitoral, quantidade de sangue no corpo de um indivíduo etc.
- **Amostra:** Subconjuntos da população.
  - É possível inferir sobre uma amostra.
    - Se for feita corretamente deve ser capaz de representar as mesmas características da população de onde foi retirada.
    - Se ela não representa a população é considerada enviesada.
- **Censo:** Pesquisa com toda a população, logo não se trabalha com amostra.
- **Enviesamento:** Subestimar ou superestimar o parâmetro da população.
  - Algumas causas para que ocorra:
    - Pesquisa de pessoas próximas ou de fácil acesso.
    - Pesquisas pela internet.
    - Sem uso de mecanismos de seleção aleatória.
- **Custo da Amostra:** 
  - Margem de Erro e Nível de Confiança
  - Variação: Amostras diferentes apresentam resultados diferentes.
  - Podemos "medir" a variação esperada.

No mundo real a amostragem é amplamente utilizada em pesquisa, estudos etc. Faz parte do dia-a-dia de um Cientista de Dados.

### Principais tipos de amostras

**1. Aleatórias Simples**
- Um determinado número de elementos é retirado da população de forma aleatória.
- Todos os elementos da população alvo do processo de amostragem devem ter as mesmas chances de ser selecionados para fazer parte da amostra.
- Pode ser feita COM ou SEM reposição.

**2. Amostra Estratificada**
- Os estratos podem ser raça, cor, religião etc.

**3. Amostra Sistemática**
- É escolhido um elemento aleatório e a partir dele a cada N elementos um novo membro é escolhido.
  - N é um intervalo que será padronizado. 
  - N é de livre escolha, pode ser 5, 10, 100 etc.


# Medidas de Centralidade e Variabilidade

### Média

- Média Amostral

$
\bar X = \frac{\sum X}{n}
$

- Média Populacional

$
\mu = \frac{\sum X}{N}
$

### Média Ponderada
- Cada número está sujeito a um peso.

$
\bar p = \frac{x_1p_1 + ... x_np_n}{p_1+ ... +p_n}
$

Exemplo:
- Prova A
  - Peso 1
  - Nota 7
- Prova B
  - Peso 2
  - Nota 9

$
\bar p = \frac{7 * 1 + 9*2}{1*2} = 8.3
$

### Moda
- Número mais frequente.

### Mediana
- É literalmente o valor do meio da base de dados.
- Como se calcula:
  - Ordena-se os valores em ordem crescente.
  - Se a quantidade for par, a mediana será a média dos valores nas posições $\frac{n}{2}$ e $\frac{n}{2}+1$.
  - Se a quantidade for impar, a mediana será o valor na posição $\frac{n+1}{2}$.

**Média VS Mediana**
- Dados: 10, 20, 30, 40, 10.000.
- Média: 2.020 - Influenciada pelo outlier.
- Mediana: 30. - Sofre menos da influência.

### Variância

- Junto com o desvio padrão, mostram a regularidade dos dados, ou seja, como os dados variam com relação à média.
- Não está relacionada à escala dos dados, logo independe de ter números altos ou baixos.
- Mostram como os dados variam entre si.

- Variância Populacional

$
\sigma^2 = \frac {\sum (x_i - \bar x)^2}{N}
$

- Variância Amostral

$
s^2 = \frac {\sum (x_i - \bar x)^2}{n - 1}
$

### Desvio Padrão

- Quanto maior, mais os dados estão afastados da média.
- Se igual a zero os dados são todos iguais, ou seja, não há variação.

- Desvio Padrão Populacional

$
\sqrt \sigma^2 = \sigma
$

- Desvio Padrão Amostral

$
\sqrt s^2 = s
$

### Amplitude

- Maior Valor - Menor Valor

### Medidas Não Centrais

1. Quartis
  - Q1: 25% dos menores valores
  - Q2: 50% dos menores valores(equivale a mediana)
  - Q3: 75% dos menores valores
Há também o Q0 (menor valor) e Q4 (maior valor), mas dificilmente são utilizados com essa notação.

### Quando utilizar o Desvio Padrão e quando a Variância

- Deve-se sempre utilizar os dois.
  - Esse é o ideal.
- A variância é influenciada por valores muito distantes.
- Pode-se analisar a variação destacando ou não discrepâncias (variância, desvio padrão).

# Probabilidade
- Basicamente é chance de algo ocorrer ou não. 
  - Exemplo a chance de ganhar na loteria.
- Conceitos:
  - A probabilidade (P) sempre está entre o intervalo de 0 $\leq$ P $\leq$ 1.
  - P = 1: Evento certo.
  - P = 0: Evento impossível.
  - Probabilidade pode ser representada das seguintes formas:
    - 0.5
    - 50%
    - $\frac{1}{2}$

- Algumas impossibilidades:
  - -0.5
  - -20%
  - $\frac{2}{1}$

## Conceitos

- Experimento: O que está sendo estudado.
  - Exemplo: Jogar Moeda.
- Espaço Amostral: Todas as possibilidades de ocorrência do evento.
  - Exemplo: Cara ou Coroa.
- Evento: Resultados ocorridos.
  - Exemplo: Dar cara.

## Analisando os eventos

1. Eventos excludentes:
- Quando não podem ocorrer ao mesmo tempo.
  - Exemplo: Dar um e par ao mesmo tempo.
2. Eventos NÃO excludentes:
- Quando podem ocorrer ao mesmo tempo.
  Exemplo: Jogar um dado e dar 2 e par ao mesmo tempo.

3. Eventos Dependentes:
- A ocorrência de um afeta o outro. Um ocorre e para depois o outro ocorrer.
  - Exemplo: Tirar 2 cartas sem reposição, ao tirar a primeira carta, as chances das demais alteram para a próxima retirada.

4. Eventos Independentes:
- Um não evento altera o outro.
  - Jogar o dado mais de uma vez.

## Como se calcula a probabilidade de um evento?

$
P = \frac {Ocorrência\,Esperada}{Número\,de\,Eventos\,Possíveis}
$

<br />


Exemplo 1: 

Jogar uma moeda e dar cara.

## Eventos Excludentes

- O cálculo é feito somando as probabilidades.

Exemplo 1:

Jogar um dado e ser 1 ou par.

<br />


$
P = \frac {1}{6} +  \frac {3}{6}
$

<br />


$
P = \frac {4}{6}
$

<br />

$
P = 0.66
$

<br />
<br />

## Eventos Não Excludentes

- O cálculo é feito somando as probabilidades e diminuindo as sobreposições.

Exemplo 1:

Jogar um dado e ser 2 ou par.

<br />


$
P = \frac {1}{6} +  \frac {3}{6} - \frac {1}{6}
$

<br />


$
P = \frac {3}{6}
$

<br />

$
P = 0.5
$

<br />

## Eventos Independentes

- O cálculo é feito multiplicando as probabilidades dos eventos.

Exemplo 1:

Jogar dois dados e dar 1 e (dois eventos indepentes).


<br />


$
P = \frac {1}{6} *  \frac {1}{6}
$

<br />


$
P = \frac {1}{36}
$

<br />

$
P = 0.027
$

<br />

## Eventos Dependentes

- O cálculo é feito multiplicando as probabilidades dos eventos, porém tem que se atentar à nova probabilidade das.

Exemplo 1:

Com 6 cartas na mão (A,2,3,4,5,6), qual a probabilidade do primeiro evento tirar A e no segundo tirar 4? (No segundo evento temos 1 carta a menos no baralho)


<br />


$
P = \frac {1}{6} *  \frac {1}{5}
$

<br />


$
P = \frac {1}{30}
$

<br />

$
P = 0.033
$

# Estocástico x Determinístico

# Probabilidades em Jogos de Azar

- As dicas mais comuns são:
  - Jogar números que saem mais.
  - Jogar número que saem menos.
  - Eliminar combinações sem pares ou sem impares.

- Sendo os sorteios **eventos independentes**, as probabilidades de sorteio são as mesmas.
  - 1, 2, 3, 4, 5, 6.
  - Números favoritos.
  - Outra combinação
  - etc.
Todas têm a mesma chance.

A mesma coisa de jogar um dados diversas vezes, por serem independentes, todas as vezes terão a mesma chance.

### Motivo de nunca ter saido 1, 2, 3, 4, 5, 6 na Mega-Sena

- São 50.036.860 combinações possíveis.
2 vezes por semana = 104 sorteios por ano.
- As chances são sempre as mesmas, ainda que não fossem repetidos, demorariam 418.123 anos para sair todos os possíveis.
