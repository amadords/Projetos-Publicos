**Importação da bibliotecas**

```
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

**Dados utilizados**

Vamos criar duas listas, uma com os dados de antes (primavera) e uma com os dados de depois (verão).

Uma terceira lista terá apenas o nome das familias olfativas que estarão na apresentação.

```
# dados a serem plotados
labels = ['Orientais', 'Amaderados', 'Frutais', 'Cítrico', 'Floral']
antes = [7, 6, 29, 23, 35]
depois = [5, 7, 20, 27, 41] 
x = np.arange(len(labels)) # criando mapeamento de 0 a 4
width_bar=0.35 # tamanho das barras
```

## Gráfico padrão

Abaixo criaremos um gráfico que podemos chamar de gráfico padrão de apresentação.

Veja que tem o nome das estações e um gráfico de antes e depois onde conseguimos entender o que se passa de forma fácil.

Talvez a única coisa que há de diferente são os *rects*, mas basicamente eles dividem em dois cada área do gráfico. O antes é o x - e o depois fica no x +.

```
# criar uma figura para grafico de barras, orientado a objeto
fig, ax = plt.subplots(figsize=(8, 4)) # tamanho da figura

# plotar antes e depois saparadamente
# o width_bar será dividido por 2 que um seja plotado na primeira metade e outro na segunda
rects1 = ax.bar(x - width_bar/2, antes, width_bar, label='Primavera')
rects2 = ax.bar(x + width_bar/2, depois, width_bar, label='Verão')

# labels do eixo
ax.set_xticks(x) # se for colocado, não plotará todos os nomes
ax.set_xticklabels(labels) # nome dos labels em x
ax.set_ylabel('Porcentagem') # titulo de Y

# título
ax.set_title('Volume de Vendas')

# legenda
ax.legend()

# plotar gráfico
fig.tight_layout() # para ajustar os espaçamentos
plt.savefig('datastorytelling-grafico-padrao.png',format='png')
plt.show()
```

## Gráfico para Storytelling

Trabalha com técnicas de psicologia, cognição, negócio etc.

Antes de tudo vamos mudar a resolução do gráfico.

O *svg* transforma em uma imagem vetorial.

```
%matplotlib inline
%config InlineBackend.figure_format = 'svg' 
```

Para um melhor storytelling é melhor uma fonte não serifada e é isso o que faremos.

Definiremos também as cores em formato hexadecimal.

As cores verdes e cinza utilizaremos agora e os azuis usaremos posteriormente também com o cinza.




```
# configurando parâmetros das fontes
plt.rcParams['font.family'] = 'sans-serif'

# definindo cores do gráfico em sistema hexadecimal
CINZA1, CINZA2, CINZA3= "#CFCDCE", "#A2A1A3", "#787878"
AZUL1, AZUL2 = "#8db7eb", "#253760"
VERDE1, VERDE2 = '#7ecfc9','#04f7e3'
```

Basicamente estamos criandos as mesmas configurações de um gráfico padrão, contudo faremos algumas alterações.

1. Removemos os grids, ou seja, as linhas de contorno do gráfico.
2. Removemos os títulos do gráfico e legenda. Há uma máxima que se seu gráfico precisa de legenda ele não está transmitindo o que precisa ser transmitido.
3. Criamos título e subtítulo e as anotações próximo à área do gráfico onde está nossa informação importante.
4. Alteramos as cores para quando alguém olhar, ter a atenção chamada diretamente para o que importa (lado direito do gráfico).

```
# ajustes para construir o gráfico
x = np.arange(len(labels))  # criando mapeamento de 0 a 4
width_bar = 0.35 # tamanho das barras

# definindo cores por plot
cores_antes = [CINZA1, CINZA1, CINZA1, CINZA3, CINZA3]
cores_depois = [VERDE1, VERDE1, VERDE1, VERDE2, VERDE2]
cores_depois2 = [AZUL1, AZUL1, AZUL1, AZUL2, AZUL2]
```

```
# criando uma figura para grafico de barras
fig, ax = plt.subplots(figsize=(8, 4))

# plotar antes e depois saparadamente
# o width_bar será dividido por 2 que um seja plotado na primeira metade e outro na segunda
rects1 = ax.bar(x - width_bar/2, antes, width_bar, label='Antes', color=cores_antes)
rects2 = ax.bar(x + width_bar/2, depois, width_bar, label='Depois', color=cores_depois)

# labels do eixo
ax.set_xticks(x)
ax.set_xticklabels(labels)
# ax.set_ylabel('Porcentagem')

# # removendo grids e eixos (linhas)
ax.spines['right'].set_visible(False) # direita
ax.spines['left'].set_visible(False) # esquerda
ax.spines['top'].set_visible(False) # em cima
ax.spines['bottom'].set_visible(False) # embaixo

# # removendo eixo y
plt.gca().axes.get_yaxis().set_visible(False)

# # removendo xticks do eixo x
plt.tick_params(
    axis='both',         
    which='both',     
    bottom=False,      
    top=False,         
    labelbottom=True) # já é o padrão # passar False caso queira retirar os nomes 

# função para criar labels automaticamente com %
def autolabel(rects):
    for rect in rects:                                                   # dentro dos nossos rects
        height = rect.get_height()                                       
        ax.annotate(                                                     # anotação para cada label
                    '{}%'.format(height),                                # formato com %
                    xy=(rect.get_x() + rect.get_width()/2,  height),     # para fazer o descolamento
                    xytext=(0, -20),  # offset                           # esse 
                    textcoords="offset points",                          # e esse são para localização
                    ha='center',                                         # para alinhar centralizado na horizontal
                    va='bottom',                                         # para alinhar verticalmente # não precisa nesse caso por já estar
                    fontsize=10,                                         # tamanho da fonte
                    color="white",                                       # cor da fonte
                    )
        
# plotando labels dentro das barras com a função
autolabel(rects1)
autolabel(rects2)

# plotando título
plt.text(-0.35, 52, 'Percentual de Vendas',
         fontsize=16, 
         color=CINZA3,
         weight='bold')

# plotando subtitulo
plt.text(-0.35, 49,
         "Com base na familia olfativa dos\n"
         "perfumes nas estações Primavera e Verão",
         verticalalignment="top",
         fontsize=12,
         color=CINZA2)

# plotando anotações do resultado da campanha
plt.text(4.5, 39,
         "$\\bf{RESULTADO}$\n"
         "No verão os\n"
         "perfumes cítricos\n"
         "e florais se tornam\n"
         "os mais comprados.\n"
         ,
         verticalalignment="top",
         fontsize=12,
         color=VERDE2)

# plotando o gráfico
fig.tight_layout()
# se o savefig for passado após plotar o gráfico, a figura é zerada
plt.savefig('datastorytelling.png',format='png')
# plotando
plt.show()
```

Veja que, quando você bate o olho no gráfico, você é remetido diretamente para o lado direito do gráfico e era o que eu queria, pois é onde está aquilo que é importante que você saiba.

Veja que o lado esquerdo do gráfico está ofuscado e o título não está com aquele preto forte que exige esforço cognitivo da sua parte.

Na verdade, todo o gráfico reforça isso.

Baseado no gráfico anterior (Gráfico padrão), lá dá para perceber que, olhando, não dá para saber qual a informação passada tão facilmente como aqui.

Isso faz muita diferença, contudo há um trabalho adicional, logo não será utilizado todo tempo, mas em momentos cruciais dá para fazer e deve ser feito.
